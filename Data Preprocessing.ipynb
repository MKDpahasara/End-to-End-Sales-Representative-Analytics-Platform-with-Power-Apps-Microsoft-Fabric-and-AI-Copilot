```python
df = spark.sql("SELECT * FROM Sales_Rep_Data_lakehouse.SalesRepData LIMIT 1000")
display(df)
```

```python
df.head()
```

```python
df.printSchema()
```

```python
df.describe().show()
```

```python
df.toPandas().head()
```

```python
df = df.select(
    "Title",
    "Visit_Date",
    "Sales_Rep_ID",
    "Sales_Rep_Name",
    "Area",
    "Shop_Name",
    "Target_Sales_Limit",
    "Target_Achieved"
)
```

```python
from pyspark.sql.functions import when, col

df_viz = df.withColumn(
    "Target_Achieved",
    when(col("Target_Achieved") == 1.0, "Yes")
    .when(col("Target_Achieved") == 0.0, "No")
)
```

```python
df_viz.toPandas().head()
```

```python
from pyspark.sql.functions import to_date, col

df = df.withColumn(
    "Visit_Date",
    to_date(col("Visit_Date"))
)
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn(
    "Title",
    to_timestamp(col("Title"), "HH:mm:ss")
)
```

```python
from pyspark.sql.functions import to_date

df = df.withColumn(
    "Visit_Date",
    to_date(col("Visit_Date"))
)
```

```python
spark.conf.set("spark.sql.legacy.timeParserPolicy", "LEGACY")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.toPandas()
```

```python
sns.barplot(data=df, x="Area", y="Target_Sales_Limit")
plt.title("Average Target Sales Limit by Area")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x="Target_Achieved")
plt.title("Target Achievement Distribution")
plt.show()
```

```python
sns.histplot(data=df, x="Target_Sales_Limit", bins=10, kde=True)
plt.title("Distribution of Target Sales Limit")
plt.show()
```
